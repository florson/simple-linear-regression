Simple linear regression example without using scikit-learn
(least squares method)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/mubaris/potential-enigma/master/headbrain.csv')
```

```python
data.head()
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
```

```python
n = len(X)
n
```

```python
Sx = X.sum()
Sy = Y.sum()
Sxx = sum(map(lambda i: i*i, X))
Syy = sum(map(lambda i: i*i, Y))
Sxy = sum(X*Y)
delta = n*Sxx - (Sx)**2
```

```python
a = (n*Sxy - Sx*Sy) / delta
a
```

```python
b = (Sxx*Sy - Sx*Sxy) / delta
b
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100
```

```python
x = np.linspace(min_x, max_x, num = 1000)
y = a*x + b
```

```python
plt.plot(x, y, c='#58b970', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
```

Calculating Root Mean Squares Error

```python
rmse = 0
for i in range(n):
    y_pred = a*X[i] + b
    rmse += (Y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/n)
print(rmse)
```
